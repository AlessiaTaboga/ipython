```python
# There are two syntax styles of using matplotlib: objected oriented or MATLAB
```

```python
# Matlab style:

plt.figure()
# create two panels and set axis
plt.subplot(2,1,1)  # rows, columns, panel numbers of my subplot grid
plt.plot(x, np.sin(x))

# create the second panel and set axis
plt.subplot(2,1,2) # rows, columns, panel numbers of my subplot grid
plt.plot(x,np.cos(x));
```

```python
# above we say that we want to do a figure: we create an active context
# then we say that the context is a subplot (rows, columns, panel)
# we say that we are in the first panel
# plot x and np.sin(x)

# then we modify the context by drawing the second plot
# (Matlab style)
# Limitations: we need to keep track of the active context and we cannot go back
```

```python
# Exercise:
# plot 3 graphs next to eachothers with the third showing sin+cos
plt.figure()
# create 1st panel and set axis
plt.subplot(1,3,1)  # rows, columns, panel numbers of my subplot grid
plt.plot(x,np.sin(x))

# create the second panel and set axis
plt.subplot(1,3,2) # rows, columns, panel numbers of my subplot grid
plt.plot(x,np.cos(x));

# create the third panel and set axis
plt.subplot(1,3,3) # rows, columns, panel numbers of my subplot grid
plt.plot(x,np.cos(x)+np.sin(x));
```

```python
# Generate the same thing using object oriented syntax:
# create grid of plots,
# ax will be an array of the Axes objects
fig, ax = plt.subplots(2)

# call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
# we don't have an active context.
# we create a grid object (figure,axis - ax here means both x and y)
# this is more convenient compared to the Matlab style
```

```python
# if I want to add another plot now

fig, ax = plt.subplots(3)

# call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
ax[2].plot(x, np.sin(x)+np.cos(x))
```

```python
# the Matlab syntax is more practical for quick visualisation, but for more complicated visualisations,
# its stateful interface makes it easy to run into problems. For example, once a panel is created,
# you can't go to the previous one and modify it. Therefore the object-oriented interface is considered more standard
# for complicated visualisation.
```

```python
# we want to modify it
# COLOURS:
plt.plot(x,np.sin(x-0), color='blue') # specify colour by name
plt.plot(x,np.sin(x-1), color='m')    # short colour code (rgbcmyk)
plt.plot(x,np.sin(x-2), color='.75') # Grayscale between 0 and 1
plt.plot(x,np.sin(x-3), color='#FFDD44')  # Hex code (RRGGBB from 00 to FF) HTML colours
plt.plot(x,np.sin(x-4), color=(1.0,0.2,0.3))  # RGB tuple, values 0 to 1 
plt.plot(x,np.sin(x-5), color='black')  # HTML colour name
```

```python
# the main difference between plt.scatter and plt.plot is that
# it can be used to create scatter plots where the properties of each individual point
# (size, face color, edge color, etc) can be individually controlled or mapped to data
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x,y, c=colors, s=sizes, alpha=0.7, cmap ='viridis')
plt.colorbar();
```

```python
# randn(100)= 100 random numbers normally distrubuted
# alpha = transparency
# cmap = color palette
# in colors we created random numb from 0 to 1 - were 0 and 1 are the extremes of the palette
# above we are setting random x and y coordinates
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.6, s=100*features[3], c=iris.target, cmap='inferno')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

```python
# flowers (iris) were collected by a scientist
# he noted the sepal length, sepal width and petal length
# the flowers were groupped in 3 classes (above yellow, red, black)
# He wanted to come up with a classification method
# the scatter plot helps us identifying how we can separate the various classes
# we can see easily that there is a line that separates the black from the others
# for the scatter plot from iris dataset we used
# feature[0]= sepal length
# feature[1]= sepal width
# feautre[3]= size
# target (type of flower) was used for color c=iris.target
```
